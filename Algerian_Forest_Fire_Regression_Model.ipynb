## Problem statement: We have to predict temperature based on other features

So, here our temperature feature is the dependent feature and rest of all are independent features

```python
# Dependent feature
y = data['Temperature']
```

```python
# Independent features
X = data.drop('Temperature',axis=1)
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=26)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
## Standardize or feature scaling the datasets
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model Training : Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
## Coefficients
regression.coef_
```

```python
## Intercept
regression.intercept_
```

```python
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

## Assumptions

```python
plt.scatter(y_test,y_pred)
```

Original data and Predicted data should have some linear relationship 

```python
residuals = y_test - y_pred
```

```python
residuals
```

```python
sns.distplot(residuals,hist=False)
```

Residuals should follow a Normal Gaussian Distribution

```python
## Residuals and predicted values should be uniformly distributed
plt.scatter(y_pred,residuals)
```

```python
## Check the perform metrics 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
```

```python
print("MSE: ",mean_squared_error(y_test,y_pred))
```

```python
print("MAE: ",median_absolute_error(y_test,y_pred))
```

```python
print("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

## R-Squared and Adjusted R-Squared for model performance

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
```

```python
score
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
